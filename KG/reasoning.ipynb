        P -> Q  
        P  
        Q
    2. 否定后件推理  
        P -> Q  
        ¬Q  
        ¬P  
    3. 三段论  
        P -> Q

        P -> R

- 类比推理  
    个体到个体的推理过程

# 在知识图谱上实现推理
---
- 知识图谱将推理任务进行了简化，并将推理的任务限定于基于知识图谱的图的结构，完成关系的推理和事实的预测  
- 知识图谱的每一条三元组都代表了关于客观世界的事实描述，同时图的结构进一步为推理提供了关联线索  
    知识图谱的目标是利用图谱中已知的关联关系或事实推断未知的关系或事实

# 常见知识图谱推理方法分类  

<h3>从推理模式的维度可以划分为演绎推理和归纳推理</h3>  

- 演绎推理更多依赖于显式的知识描述和逻辑推导  
- 归纳推理更多依赖于大数据和机器学习方法  

<img src="../img/5.png">
  

## 基于符号表示的推理
---
- Tbox语言  
    1. 定义：引入概念以及关系的名称  
        e.g:Mother, Person, has_child  
        
    2. 声明包含关系的公理  
        e.g:Mother ⊆ ∃ has_child.Person
- Abox语言(事实性断言或描述)
    1. 概念新言——表示一个对象是否属于某个概念  
        e.g:Mother(Alice), Person(Bob)  
        
    2. 关系新言——表示两个对象是否满足特定关系  
        e.g:has_child(Alice, Bob)  
---
演绎推理不存在可解释性的问题  
显式的路径特征或规则可直接用于对推理结果进行解释

## 基于向量表示的推理
---
- 将知识图谱中的事件和关系都投影到向量空间，再通过向量空间完成计算  
    通常以三元组为训练输入，通过定义一个约束函数，对向量表示进行学习
- 将表示向量化和参数化的后果是丢失可解释性
- 图神经网络可以为图谱推理获得更多的图结构方面的特征，但图谱稀疏性问题更加突出

# 符号逻辑与表示学习的融合
---
- 符号知识的向量化表示，更易于捕获隐含的不确定性知识，向量空间的连续性，也使得基于向量的推理计算更易于泛化，并免于严格的符号逻辑匹配  
- 通过可微过程学习出的数字化知识表示，对机器友好，但人不可理解;  
  向量化的推理计算丢失了对推理面的跟踪，因而导致推理结果不可解释
---
<h3>将符号规则和表示学习相结合,可以一定程度上解决这一问题</h3>  

---
- 向量表示的学习过程中，植入符号规则的学习过程，我们可以利用同步学术的规则，增强推理结果的可解释性  
- 人工构建和机器学习出的符号规则与向量表示学习相耦合，也可以进一步在表示学习过程中植入逻辑层的知识鲜艳，从而使得向量表示更易于实现符号逻辑推理
<img src="../img/6.png">  
- 符号逻辑与表示学习的互补融合是解决表示学习可解释性的一个重要途径
- 符号逻辑的向量化和神经网络化也使得感知层的任务，比如抽取，识别，人和认知层的任务，比如理解和推理，能够在统一的表示空间完成，因此也是感知认知进一步融合的未来的主要发展趋势
