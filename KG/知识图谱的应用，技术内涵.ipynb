# 知识图谱的应用
- 互联网搜索引擎（事物级别而非文本级）
- 智能问答（对话式搜索）
    1. 问答对（优：易于管理 缺：无法进行精确回答）
    2. 给定问句，从大量文本中准确定位答案
    3. 知识图谱
- 推荐系统（挖掘更深层次用户兴趣，实现更加个性化推荐）
- 大数据分析
- 语言理解
- 视觉理解
- 语义辅助设备互联
---
知识图谱技术源于互联网， 最早落地应用是<font color=Red>搜索引擎、智能问答和推荐计算</font>等技术领域  

知识图谱通过规范化语义融合<font color=Red>多来源数据</font>，并能通过<font color=Red>图谱推理</font>能力支持复杂关联数据的挖掘分析，因此在<font color=Red>大数据分析领域</font>也有广泛应用 

不论是语言理解和视觉理解，外源知识库的引入都可以有力的提升语义理解的深度和广度  

知识图谱在<font color=Red>医疗、金融、电商、通信</font>等多个垂直领域都有广泛的应用，并且在每个领域都有独特的实现和实践方式

## 技术内涵：
- 基于图的知识表示  
    有向标记图：最简单、接近自然语言和人脑认知的数据模型  
    - RDF为代表 （subject， predicate， object）  
- 图数据库存储与查询  
    - 图数据库充分利用图的结构建立微索引，这种微索引比起关系数据库的全局索引在处理图遍历查询时更加廉价，其查询复杂度与数据集整体大小无关，仅正比于相邻子图的大小。因此在很多涉及复杂关联和多跳的场景中得到广泛应用。  
- Knowledge Base Population（知识抽取）  
    <img src="../img/15.png" align=right>  
    
    - 从不同来源、不同结构的数据中进行知识抽取，形成知识存入到知识图谱  
    - 文本一般不作为知识图谱构建的初始来源，而多用来做知识图谱的补全  
- 知识图谱融合  
    <img src="../img/16.png">  
- **知识图谱推理（核心技术和任务）**  
    - 目标：利用知识图谱中已经存在的关联关系或事实推断未知的关联关系或事实  
    - 推理可以实现链接预测，补全缺失的属性，检测错误的描述，识别语义冲突，以提升图谱的质量  
    - 在查询和问答中，推理可以用来拓展问句的语义，提高查询的召回  
    - 在推荐计算中，推理可以提高推荐的精准性和可解释性  
    - 深度语言语义理解和视觉问答中也扮演必不可少的角色  
- 知识图谱问答——KBQA
    - 处理流程：对问句的语义解析（自然语言解析为更容易被机器所处理的逻辑表示或分布式表示形式）-->将问句中的语义表示与知识图谱中的结点进行匹配、查询、推理-->对候选的匹配结果进行排序，生成对用户友好的答案形式  
- 图算法与图神经网络  
    - 各种常用的图表式学习技术都可以用来对知识图谱作深入的挖掘与分析
